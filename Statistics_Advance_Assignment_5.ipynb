# Q5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation
of 12 using Python. Interpret the results.

To calculate the 90% confidence interval for a sample with a given mean and standard deviation, you can use the following formula:

[
\text{Confidence Interval} = \text{mean} \pm z \times \left(\frac{\text{standard deviation}}{\sqrt{n}}\right)
]

Where:

* ( z ) is the z-score corresponding to the desired confidence level (for 90%, ( z \approx 1.645 )).
* ( \text{mean} ) is the sample mean.
* ( \text{standard deviation} ) is the sample standard deviation.
* ( n ) is the sample size.
If we do not have the sample size ( n ), we can't calculate the margin of error due to the standard error term (\frac{\text{standard deviation}}{\sqrt{n}}). Let's assume ( n = 30 ) for this calculation (you can adjust it for your actual sample size if it's different).


```python
import scipy.stats as stats
import numpy as np

# Given data
mean = 75
std_dev = 12
n = 30  # Assuming a sample size of 30 for this example

# Calculate the z-score for 90% confidence
z = stats.norm.ppf(0.95)  # For a 90% CI, 5% in each tail

# Calculate the standard error
standard_error = std_dev / np.sqrt(n)

# Calculate the margin of error
margin_of_error = z * standard_error

# Calculate confidence interval
confidence_interval = (mean - margin_of_error, mean + margin_of_error)

confidence_interval
```

# Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the
area corresponding to a chi-square statistic of 15.

To plot the chi-square distribution with 10 degrees of freedom and shade the area corresponding to a chi-square statistic of 15, we can use the matplotlib and scipy libraries in Python.

Here is how to do it step by step:

1. Import the necessary libraries.
2. Create the x values representing the range of the chi-square distribution.
3. Compute the chi-square distribution values.
4. Plot the distribution.
5. Shade the area corresponding to a chi-square statistic of 15.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parameters
df = 10  # degrees of freedom
x = np.linspace(0, 30, 1000)  # x values for the chi-square distribution

# Chi-square distribution
y = stats.chi2.pdf(x, df)

# Plotting the Chi-square distribution
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Chi-square Distribution (df={df})', color='blue')

# Shading the area corresponding to a chi-square statistic of 15
x_fill = np.linspace(15, 30, 1000)
y_fill = stats.chi2.pdf(x_fill, df)
plt.fill_between(x_fill, y_fill, color='lightblue', alpha=0.5, label='Area for Chi-square > 15')

# Vertical line at chi-square = 15
plt.axvline(15, color='red', linestyle='--', label='Chi-square = 15')

# Labels and title
plt.title('Chi-Square Distribution with 10 Degrees of Freedom')
plt.xlabel('Chi-Square Value')
plt.ylabel('Probability Density Function (PDF)')
plt.legend()
plt.grid()

# Show the plot
plt.show()
```

# Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520
preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who
prefer Coke.

To calculate a 99% confidence interval for the proportion of people who prefer Coke based on a sample of 1000 people, we can use the formula for the confidence interval for proportions:

[
\text{Confidence Interval} = \hat{p} \pm z \times \sqrt{\frac{\hat{p}(1 - \hat{p})}{n}}
]

Where:

* (\hat{p}) is the sample proportion (in this case, the proportion that prefers Coke).
* (z) is the z-score that corresponds to the desired confidence level (for a 99% confidence level, (z \approx 2.576)).
* (n) is the sample size.
# Step-by-step Calculation:
1. Calculate the sample proportion ((\hat{p})):

* Sample size (n = 1000)
* Number of people who prefer Coke = 520
* (\hat{p} = \frac{520}{1000} = 0.52)
2. Calculate the standard error (SE):
[
SE = \sqrt{\frac{\hat{p}(1 - \hat{p})}{n}} = \sqrt{\frac{0.52(1 - 0.52)}{1000}} = \sqrt{\frac{0.52 \times 0.48}{1000}} \approx 0.016
]

3. Find the z-score for a 99% confidence level:
[
z \approx 2.576 \quad (\text{from z-tables for 99% CI})
]

4. Calculate the margin of error (ME):
[
ME = z \times SE = 2.576 \times 0.016 \approx 0.041
]

5. Calculate the confidence interval:
[
\text{Confidence Interval} = \hat{p} \pm ME = 0.52 \pm 0.041
]
[
\text{Lower Bound} = 0.52 - 0.041 = 0.479
]
[
\text{Upper Bound} = 0.52 + 0.041 = 0.561
]



```python
import scipy.stats as stats
import numpy as np

# Given data
n = 1000  # sample size
x = 520   # number of people who prefer Coke
p_hat = x / n  # sample proportion

# Standard error calculation
se = np.sqrt(p_hat * (1 - p_hat) / n)

# z-score for 99% confidence interval
z = stats.norm.ppf(0.995)  # 0.995 because it's two-tailed

# Margin of error
margin_of_error = z * se

# Confidence interval calculation
confidence_interval = (p_hat - margin_of_error, p_hat + margin_of_error)

confidence_interval
```

# Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe
45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the
expected frequencies of a fair coin. Use a significance level of 0.05.

To conduct a chi-square goodness of fit test to determine whether the observed frequencies of coin flips match the expected frequencies of a fair coin, we can follow these steps:

1. efine the Null and Alternative Hypotheses:

* Null Hypothesis ((H_0)): The coin is fair; the probabilities of getting heads or tails are equal (0.5 for each).
* Alternative Hypothesis ((H_a)): The coin is not fair; it is biased towards tails.
2. Collect Data:

* Total flips ((n)) = 100
* Observed tails (O) = 45
* Therefore, observed heads (H) = (100 - 45 = 55)
3. Calculate the Expected Frequencies:
For a fair coin, the expected frequency of heads and tails for 100 flips would be:
[
E = \text{Total Flips} \times P(\text{Heads or Tails}) = 100 \times 0.5 = 50 \text{ (for both heads and tails)}
]

4. Calculate the Chi-Square Statistic:
The formula for the chi-square statistic ((\chi^2)) is given by:
[
\chi^2 = \sum \frac{(O - E)^2}{E}
]
For our data:

* Observed Frequencies (O): [45 tails, 55 heads]
* Expected Frequencies (E): [50 tails, 50 heads]
[
\chi^2 = \frac{(45 - 50)^2}{50} + \frac{(55 - 50)^2}{50}
]

5. Determine the Degrees of Freedom:
The degrees of freedom for a goodness of fit test is given by:
[
df = k - 1
]
where (k) is the number of categories. In this case, we have heads and tails (2 categories):
[
df = 2 - 1 = 1
]

6. Find the Critical Value:
Using a chi-square distribution table or calculator, we find the critical value for (\alpha = 0.05) and (df = 1).

7. Make a Decision:
If the calculated (\chi^2) statistic exceeds the critical value from the table, we reject the null hypothesis.


```python
import scipy.stats as stats

# Given data
n = 100  # total flips
observed = [45, 55]  # observed tails and heads
expected = [n * 0.5, n * 0.5]  # expected for a fair coin

# Calculate chi-square statistic
chi_square_statistic = sum((o - e) ** 2 / e for o, e in zip(observed, expected))

# Degrees of freedom
df = len(observed) - 1

# Critical value at 0.05 significance level
alpha = 0.05
critical_value = stats.chi2.ppf(1 - alpha, df)

# Output results
print("Chi-Square Statistic:", chi_square_statistic)
print("Critical Value (alpha = 0.05):", critical_value)

# Making a decision
if chi_square_statistic > critical_value:
    print("Reject the null hypothesis: The coin is biased.")
else:
    print("Fail to reject the null hypothesis: No evidence of bias.")
```

# Q11. A random sample of 30 people was selected from a population with an unknown mean and standard
deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10.
Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a
significance level of 0.05.

To determine if the population mean is significantly different from 70 using the given sample data, we will conduct a hypothesis test. We will use a t-test for this purpose, as the population standard deviation is unknown and the sample size is relatively small (n < 30).

# Step 1: State the Hypotheses
We will define our null and alternative hypotheses as follows:

* Null Hypothesis ((H_0)): The population mean ((\mu)) is equal to 70.
[
H_0: \mu = 70
]

* Alternative Hypothesis ((H_a)): The population mean ((\mu)) is not equal to 70.
[
H_a: \mu \neq 70
]

# Step 2: Set the Significance Level
We use a significance level ((\alpha)) of 0.05.

Step 3: Calculate the Test Statistic
Given:

* Sample size ((n)) = 30
* Sample mean ((\bar{x})) = 72
* Sample standard deviation ((s)) = 10
We use the t-statistic formula:
[
t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}}
]
where:

* (\mu_0) is the population mean under the null hypothesis (70).
Substituting the values into the formula:
[
t = \frac{72 - 70}{10 / \sqrt{30}} = \frac{2}{10 / \sqrt{30}} = \frac{2 \cdot \sqrt{30}}{10}
]

Calculating (\sqrt{30}):
[
\sqrt{30} \approx 5.477
]

Now substituting it back into the equation:
[
t = \frac{2 \cdot 5.477}{10} \approx \frac{10.954}{10} \approx 1.0954
]
So:
[
t \approx 1.095
]

# Step 4: Determine the Critical Values
Since this is a two-tailed test, we need to find the critical t-values for (n-1) degrees of freedom at a significance level of 0.05. The degrees of freedom ((df)) are:
[
df = n - 1 = 30 - 1 = 29
]

Using a t-table or calculator, we find the critical t-values for (df = 29) at ( \alpha = 0.05):

* Critical t-values for a two-tailed test: ( t_{0.025, 29} ) and ( t_{0.975, 29} ) are approximately ( \pm 2.045 ).
# Step 5: Compare the Test Statistic with the Critical Values
We reject the null hypothesis if the absolute value of the test statistic is greater than the critical value:

[
|t| = 1.095 < 2.045
]

# Step 6: Conclusion
Since the test statistic ((1.095)) does not exceed the critical value ((2.045)), we fail to reject the null hypothesis.

Final Result
At the ( \alpha = 0.05 ) significance level, there is not enough evidence to conclude that the population mean is significantly different from 70.
