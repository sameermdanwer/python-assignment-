# Q1. How can you create a Bokeh plot using Python code?

Creating a Bokeh plot using Python is straightforward. Bokeh is a powerful visualization library that allows you to create interactive plots for web browsers. Here's a step-by-step guide on how to create a simple Bokeh plot:

```python
pip install bokeh
```

```python
# Step 1: Import the required libraries
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Step 2: Prepare the output (this will display the plot in a Jupyter Notebook)
output_notebook()

# Step 3: Create a figure object
p = figure(title="Basic Line Plot",
           x_axis_label='X-axis',
           y_axis_label='Y-axis')

# Step 4: Add a line renderer
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
p.line(x, y, legend_label="Line", line_width=2)

# Step 5: Show the plot
show(p)
```

# Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

In Bokeh, glyphs are the basic visual shapes or symbols that are used to represent data points on a plot. Glyphs can take various forms, such as circles, squares, lines, and more. Each glyph can be customized with different attributes, such as size, color, line width, and transparency.

# Adding Glyphs to a Bokeh Plot
To add glyphs to a Bokeh plot, you typically use the figure object to create a plot and then call methods to add specific glyphs, specifying the data source and any visual properties. Below are common glyphs you can use:

* circle(): For circular markers.
* square(): For square markers.
* line(): For line plots.
* triangle(): For triangular markers.
* rect(): For rectangular markers.
# Example of Using Glyphs in a Bokeh Plot

```python
# Step 1: Import the required libraries
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the output
output_notebook()

# Step 2: Create a figure object
p = figure(title="Bokeh Glyphs Example",
           x_axis_label='X-axis',
           y_axis_label='Y-axis')

# Step 3: Define sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Step 4: Add glyphs
# Adding circles
p.circle(x, y, size=10, color="blue", alpha=0.6, legend_label="Circles")

# Adding squares
p.square([1, 2, 3, 4, 5], [5, 6, 7, 2, 4], size=10, color="red", alpha=0.6, legend_label="Squares")

# Adding a line
p.line(x, y, line_width=2, color="green", legend_label="Line")

# Step 5: Show the plot
show(p)
```

# Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

Customizing the appearance of a Bokeh plot allows you to enhance the visual appeal and readability of your visualizations. You can customize various aspects of a Bokeh plot, including the axes, title, legend, colors, sizes, and more. Here’s how to do it, along with an example.

# Customizing a Bokeh Plot
1. Title Customization: You can set the title and customize its font size, color, and location.
2. Axis Customization: You can customize axis labels, tick labels, and the range of the axes.
3. Legend Customization: You can modify the legend's location, orientation, and appearance.
4. Glyph Customization: You can change the colors, sizes, and other attributes of the glyphs used in the plot.
# Example of a Customized Bokeh Plot

```python
# Step 1: Import the required libraries
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the output
output_notebook()

# Step 2: Create a figure object with customized title
p = figure(title="Customized Bokeh Plot",
           title_text_font_size='16pt',  # Title font size
           title_align='center',           # Title alignment
           x_axis_label='X-axis',
           y_axis_label='Y-axis')

# Set the title font size after creating the figure object
p.title.text_font_size = '16pt'


# Step 3: Define sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Step 4: Add glyphs with custom attributes
p.circle(x, y, size=10, color="blue", alpha=0.6, legend_label="Data Points")

# Adding a line
p.line(x, y, line_width=2, color="green", legend_label="Trend Line")

# Step 5: Customize axes
p.xaxis.axis_label_text_font_size = "12pt"  # X-axis label font size
p.yaxis.axis_label_text_font_size = "12pt"  # Y-axis label font size
p.xaxis.major_label_text_font_size = "10pt"  # X-axis tick label font size
p.yaxis.major_label_text_font_size = "10pt"  # Y-axis tick label font size

# Step 6: Customize legend
p.legend.location = "top_left"                # Legend location
p.legend.title = "Legend Title"               # Legend title
p.legend.title_text_font_size = "12pt"       # Legend title font size
p.legend.label_text_font_size = "10pt"       # Legend item font size
p.legend.border_line_width = 2                 # Legend border width
p.legend.background_fill_color = "lightgrey"  # Legend background color

# Step 7: Show the plot
show(p)
```

# Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?

# What is a Bokeh Server?
The Bokeh Server is a component of the Bokeh library that enables the creation of interactive web applications with real-time updates. It allows users to build applications where the visualizations can respond to user inputs (such as sliders, dropdowns, buttons, etc.) and change dynamically based on updates from the backend.

# Key Features of Bokeh Server
1. Interactivity: You can create interactive visualizations that respond to user inputs.
2. Real-Time Updates: You can update plots and data in real-time based on external data sources or user interactions.
3. Web-Based: Bokeh Server applications are served over HTTP, allowing users to interact with them via web browsers.
4. Python Backend: The server runs a Python script that can perform computations and generate data for the visualizations.
# Creating a Basic Bokeh Server Application

```python
pip install bokeh
```

```python
from bokeh.io import curdoc
from bokeh.layouts import column
from bokeh.models import Slider
from bokeh.plotting import figure

# Step 1: Create a figure object
p = figure(title="Interactive Bokeh Plot",
           x_axis_label='X-axis',
           y_axis_label='Y-axis')

# Initial data
x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]
line = p.line(x, y, line_width=2, color="blue")

# Step 2: Create a slider
slider = Slider(start=1, end=10, value=1, step=1, title="Scale Factor")

# Step 3: Define a callback function to update the plot
def update_plot(attr, old, new):
    scale = slider.value
    new_y = [i * scale for i in y]  # Scale the y-values based on the slider
    line.data_source.data['y'] = new_y  # Update the plot with new data

# Step 4: Attach the callback to the slider
slider.on_change('value', update_plot)

# Step 5: Organize the layout and add it to the current document
layout = column(slider, p)
curdoc().add_root(layout)
```

```python
bokeh serve --show interactive_plot.py
```

# Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?

Embedding a Bokeh plot into a web page or dashboard can be accomplished using web frameworks like Flask or Django. Here’s how you can do it for both frameworks:

# Using Flask to Embed a Bokeh Plot


```python
pip install flask bokeh
```

```python
from flask import Flask, render_template
from bokeh.plotting import figure, output_file, save
from bokeh.embed import server_document

app = Flask(__name__)

@app.route('/')
def index():
    # Generate a Bokeh plot
    plot = figure(title="Bokeh Plot", x_axis_label='X-axis', y_axis_label='Y-axis')
    plot.line([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], line_width=2, color='blue')

    # Save the plot to an HTML file
    output_file("plot.html")
    save(plot)

    return render_template('index.html', plot_script=server_document('http://localhost:5006/bokehapp'))

if __name__ == '__main__':
    app.run(debug=True)
```
