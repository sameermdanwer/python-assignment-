# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a powerful visualization library in Python that is built on top of Matplotlib. Here are five commonly used plots in Seaborn along with their uses:

1. Scatter Plot (sns.scatterplot):
* Use: A scatter plot is used to visualize the relationship between two continuous variables. Each point on the plot represents an observation in the dataset. It can help identify trends, clusters, and outliers.
2. Line Plot (sns.lineplot):
* Use: Line plots are used to display the relationship between two variables over a continuous interval. They are particularly useful for time series data, allowing for the visualization of trends over time.
3. Box Plot (sns.boxplot):
* Use: Box plots provide a graphical summary of the distribution of a dataset. They show the median, quartiles, and potential outliers. This plot is useful for comparing distributions across different categories or groups.
4. Heatmap (sns.heatmap):
* Use: Heatmaps are used to represent data in a matrix form, where values are represented by colors. They are commonly used to visualize correlation matrices or the frequency of occurrences in a two-dimensional space, making it easy to identify patterns.
5. Violin Plot (sns.violinplot):
* Use: Violin plots combine the features of box plots and density plots. They show the distribution of the data across different categories, displaying the probability density of the data at different values. This plot is useful for understanding the distribution shape and comparing multiple categories.

# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

To create a line plot using the "fmri" dataset from Seaborn, you can follow these steps. The fmri dataset contains information about brain activity and has columns for timepoints, signals, events, and regions. Here's how to load the dataset and create the specified line plot:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the 'fmri' dataset
fmri = sns.load_dataset('fmri')

# Create a line plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=fmri, x='timepoint', y='signal', hue='event', style='region', markers=True)
plt.title('Line Plot of Signal over Timepoint for Different Events and Regions')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.legend(title='Event and Region')
plt.show()
```

# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

To create box plots using the "titanic" dataset from Seaborn, you can load the dataset and then create two separate box plots: one for the age distribution across different passenger classes (pclass) and another for the fare distribution. Heres how you can do that:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the 'titanic' dataset
titanic = sns.load_dataset('titanic')

# Set up the figure for the box plots
plt.figure(figsize=(14, 6))

# Box plot for 'age' vs 'pclass'
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title('Box Plot of Age by Passenger Class')
plt.xlabel('Passenger Class (pclass)')
plt.ylabel('Age')

# Box plot for 'fare' vs 'pclass'
plt.subplot(1, 2, 2)  # 1 row, 2 columns, second subplot
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title('Box Plot of Fare by Passenger Class')
plt.xlabel('Passenger Class (pclass)')
plt.ylabel('Fare')

# Show the plots
plt.tight_layout()  # Adjusts the layout to prevent overlap
plt.show()
```

# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

To create a histogram of the price column in the "diamonds" dataset from Seaborn, while using the cut column as the hue, you can follow these steps. This allows you to visualize the distribution of diamond prices, differentiated by their cut quality. Heres how to do it:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the 'diamonds' dataset
diamonds = sns.load_dataset('diamonds')

# Create a histogram for the 'price' column, using 'cut' as hue
plt.figure(figsize=(12, 6))
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack', bins=30, kde=True)
plt.title('Histogram of Diamond Prices by Cut Quality')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.legend(title='Cut')
plt.show()
```

# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

To create a pair plot of the "iris" dataset from Seaborn, you can use the sns.pairplot function and set the hue parameter to differentiate the species of the flowers. This will create scatter plots for each pair of numerical columns and color them according to the species. Heres how to do it:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the 'iris' dataset
iris = sns.load_dataset('iris')

# Create a pair plot
plt.figure(figsize=(10, 8))
sns.pairplot(iris, hue='species', palette='Set2')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)  # Adjust title position
plt.show()
```

# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

To create a heatmap using the "flights" dataset from Seaborn, you first need to pivot the dataset to create a matrix format that represents the values you want to visualize. The "flights" dataset contains information about the number of passengers flying from one city to another over different months. Here's how you can create a heatmap for the total number of passengers for each month:



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the 'flights' dataset
flights = sns.load_dataset('flights')

# Pivot the dataset to create a matrix for the heatmap
pivot_flights = flights.pivot_table(index='month', columns='year', values='passengers', aggfunc='sum')

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(data=pivot_flights, cmap='YlGnBu', annot=True, fmt='g')
plt.title('Number of Passengers per Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```
