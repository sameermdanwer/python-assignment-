# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a comprehensive library in Python used for creating static, animated, and interactive visualizations. It is widely used for data visualization tasks and helps to generate plots and graphs that make it easier to analyze and understand data trends.

# Why is it used?
* Data visualization: Helps represent complex data in an easily understandable format.
* Customization: Offers extensive customization options for plots (colors, labels, markers, etc.).
* Integration: Works well with other libraries like NumPy and Pandas for efficient data handling and plotting.
* Publication-quality: Generates high-quality graphs suitable for publication or presentations.
* Interactive plots: Can generate interactive plots using widgets, zoom features, etc.
# Five Plots using Pyplot (a module of Matplotlib):
1. Line Plot:

Used to plot continuous data points, often used to visualize trends over time.
2. Bar Plot:

Represents categorical data with rectangular bars showing the frequency or value.
3. Scatter Plot:

Displays individual data points based on two variables, often used to show correlation or distribution.
4. Histogram:

Used for visualizing the distribution of a dataset by displaying frequencies of data intervals.
5. Pie Chart:

Represents data in a circular graph, useful for displaying proportions or percentages.
These plots are commonly used in data analysis, machine learning, and data science projects.

# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])


The subplot() function in Matplotlib is used to create multiple plots within a single figure, allowing for a structured comparison of different datasets or different visualizations of the same dataset. This is useful for visualizing relationships, trends, or patterns across multiple plots simultaneously without cluttering the output with multiple separate figures.

Here's how to create four line plots using the specified data with the subplot() function:

Code to Create Four Line Plots

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for the four line plots
x = np.array([0, 1, 2, 3, 4, 5])

# Line 1 Data
y1 = np.array([0, 100, 200, 300, 400, 500])

# Line 2 Data
y2 = np.array([50, 20, 40, 20, 60, 70])

# Line 3 Data
y3 = np.array([10, 20, 30, 40, 50, 60])

# Line 4 Data
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with subplots
plt.figure(figsize=(10, 8))

# First subplot
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, 1st subplot
plt.plot(x, y1, marker='o', color='blue')
plt.title('Line 1: Increasing Values')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()

# Second subplot
plt.subplot(2, 2, 2)  # 2 rows, 2 columns, 2nd subplot
plt.plot(x, y2, marker='s', color='orange')
plt.title('Line 2: Fluctuating Values')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()

# Third subplot
plt.subplot(2, 2, 3)  # 2 rows, 2 columns, 3rd subplot
plt.plot(x, y3, marker='^', color='green')
plt.title('Line 3: Linear Increase')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()

# Fourth subplot
plt.subplot(2, 2, 4)  # 2 rows, 2 columns, 4th subplot
plt.plot(x, y4, marker='d', color='purple')
plt.title('Line 4: Complex Fluctuation')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])


A bar plot (or bar chart) is a graphical representation of categorical data. In a bar plot, categories are represented on one axis (typically the x-axis), and their corresponding values (quantities or frequencies) are represented on the other axis (typically the y-axis). Each category is represented by a rectangular bar, where the height or length of the bar corresponds to the value of that category.

# Why is it Used?
Comparison: Bar plots are particularly effective for comparing the values of different categories. They allow viewers to quickly see which categories have higher or lower values.
Categorical Data: They are well-suited for displaying categorical data, where different categories are represented by distinct groups.
Clarity: Bar plots provide a clear visual representation of data, making it easier to interpret and analyze trends or patterns.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for the bar plots
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a figure with subplots
plt.figure(figsize=(12, 6))

# Vertical Bar Plot
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
plt.bar(company, profit, color='blue')
plt.title('Vertical Bar Plot of Company Profits')
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.grid(axis='y')

# Horizontal Bar Plot
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
plt.barh(company, profit, color='orange')
plt.title('Horizontal Bar Plot of Company Profits')
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')
plt.grid(axis='x')

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

A box plot, also known as a whisker plot, is a standardized way of displaying the distribution of data based on five summary statistics:

1. Minimum: The smallest data point excluding outliers.
2. First Quartile (Q1): The median of the lower half of the dataset (25th 2. 2. .2 percentile).
3. Median (Q2): The middle value of the dataset (50th percentile).
4. Third Quartile (Q3): The median of the upper half of the dataset (75th percentile).
5. Maximum: The largest data point excluding outliers.
# Why Use Box Plots?
* Visual Summary: They provide a visual summary of the central tendency, dispersion, and skewness of the data.
* Identifying Outliers: Box plots highlight potential outliers in the dataset, making it easier to understand data variability.
* Comparative Analysis: They are particularly useful for comparing distributions between multiple groups or datasets.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate the data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create the box plot
plt.figure(figsize=(10, 6))
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.title('Box Plot of Box 1 and Box 2')
plt.ylabel('Values')
plt.grid(axis='y')
plt.show()
```
