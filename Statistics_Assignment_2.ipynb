# Q1. What are the three measures of central tendency?

The three measures of central tendency are:

1. Mean: The average of a data set, calculated by summing all the values and then dividing by the number of values.
2. Median: The middle value of a data set when the numbers are arranged in order. If the set has an even number of values, the median is the average of the two middle numbers.
3. Mode: The value that appears most frequently in a data set. A data set can have more than one mode if multiple values have the same frequency.

# Q6. What is a Venn diagram?

A Venn diagram is a graphical representation used to show the relationships and logical connections between different sets or groups of items. It typically consists of overlapping circles, where:

* Each circle represents a set.
* The overlapping regions between the circles show the elements that are common to multiple sets.
* The non-overlapping parts represent elements that are unique to each set.

# Example:

* One circle represents Set A (e.g., students who play football).
* Another circle represents Set B (e.g., students who play basketball).
* The overlap (intersection) shows students who play both football and basketball.

# Q8. What do you understand about skewness in data?

Skewness in data refers to the measure of the asymmetry or deviation from the symmetrical, bell-shaped distribution of data (like the normal distribution). It tells us whether the data is spread more to one side of the mean than the other.

# Types of Skewness:
1. Positive Skewness (Right-Skewed):

* In a positively skewed distribution, the tail on the right side (higher values) is longer or fatter.
* Most of the data points are concentrated on the left side, with a few larger values pulling the mean to the right.
* Mean > Median > Mode.
* Example: Income distribution in many countries, where a small number of people have very high incomes.

2. Negative Skewness (Left-Skewed):

* In a negatively skewed distribution, the tail on the left side (lower values) is longer or fatter.
Most of the data points are concentrated on the right side, with a few smaller values pulling the mean to the left.
* Mean < Median < Mode.
* Example: Age at retirement, where most people retire around a typical age, but a few retire much earlier.
3. No Skewness (Symmetrical Distribution):

* In a perfectly symmetrical distribution (such as a normal distribution), the data is evenly spread out on both sides of the mean.
* Mean = Median = Mode.
